```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('macrodata.csv')
df.head()
```

```python
statistics = df.describe()
statistics
```

```python
plt.plot(df['year'], df['realgdp'], label='Real GDP', marker='o')

plt.plot(df['year'], df['realcons'], label='Real Consumption', marker='o')

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Comparison of Real GDP and Real Consumption Over the Years')

plt.legend()

plt.show()
```

```python
plt.scatter(df['unemp'], df['infl'], alpha=0.5)

plt.xlabel('Unemployment Rate')
plt.ylabel('Inflation Rate')
plt.title('Scatter Plot of Unemployment Rate vs. Inflation Rate')

plt.show()
```

```python
sns.kdeplot(df['cpi'], fill=True)

plt.xlabel('Consumer Price Index (CPI)')
plt.ylabel('Density')
plt.title('Density Plot of Consumer Price Index (CPI)')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['realinv'], df['tbilrate'], color='blue', alpha=0.7)

plt.title('Scatter Plot of Real Investment vs. Treasury Bill Rate')
plt.xlabel('Real Investment')
plt.ylabel('Treasury Bill Rate')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['pop'], marker='o', linestyle='-', color='green')

plt.title('Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['realdpi'], label='realdpi', marker='o', linestyle='-', color='blue')
plt.plot(df['year'], df['realgdp'], label='realgdp', marker='o', linestyle='-', color='green')

plt.title('Comparison between Real Personal Income and Real GDP Over the Years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()

plt.grid(True)
plt.show()
```
